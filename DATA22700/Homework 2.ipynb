```python
salary = pd.read_csv('/Users/keeganballantyne/Downloads/Current_Employee_Names__Salaries__and_Position_Titles.csv')
# Which departments are earning the highest annual salary
# Find the average annual salary for each department
# Make a bar plot for each of the the average annual salaries for each department
```

```python
salary_type = salary.dtypes
print(salary_type)
```

```python
grouped_salary = salary.groupby(['Department']).mean().reset_index().sort_values('Annual Salary')

grouped_salary

# dept_ann_salary = grouped_salary["Annual Salary"]
# dept_ann_salary.to_frame()
```

```python
count = salary[salary['Salary or Hourly'] == 'Salary'].groupby(salary['Department']).count().drop('Department',axis='columns').reset_index()
```

```python
department = grouped_salary['Department']
vals = ([])
for i in department:
    temp = count.loc[count['Department']== i]['Name']
    vals = np.append(vals, temp)
grouped_salary['Count'] = vals
```

```python
departments = grouped_salary['Department']
departments = departments.str.title()
grouped_salary['Department'] = departments
grouped_salary['Department'].unique()
```

```python
grouped_salary['Department'] = grouped_salary['Department'].replace("Mayor'S Office","Mayor's Office").replace(
    'Dais','Assets, Info, Sec'
).replace('Transportn','Transportation').replace(
    'Copa','COPA'
).replace('Oemc','OEMC').replace(
    'Animal Contrl', 'Animal Control'
).replace('Admin Hearng','Admin Hearings')
```

```python
count = grouped_salary['Count']
vals = ([])
for i in count:
    if (i >= 250):
        temp = 'Large (250+ employees)'
        vals.append(temp)
    elif (i >= 100) & (i < 250):
        temp = 'Midsize (100-249 employees)'
        vals.append(temp)
    else:
        temp = 'Small (1-99 employees)'
        vals.append(temp)
```

```python
grouped_salary['Department Size'] = vals
```

```python
fig = plt.figure(figsize = (10, 5))

from matplotlib import ticker
plt.rcdefaults()
fig, ax = plt.subplots()
sns.barplot(data=grouped_salary, y=grouped_salary['Department'],x=grouped_salary['Annual Salary'],hue=grouped_salary['Department Size'],palette='rocket',dodge=False)
plt.xticks(fontsize=12)
plt.yticks(fontsize=8)
# plt.axvline(x=98940, ymax=1,ymin=0, color='blue', linestyle='--',linewidth=3,label='Mean Across all Departments')
plt.title('Mean Salaries in Chicago Public Departments', fontsize=12)
plt.xlabel('Annual Salary (in thousands of dollars)',fontsize=12)
plt.ylabel('Department',fontsize=12)
plt.legend(loc='upper right',bbox_to_anchor =(1.57,1))
ax.xaxis.set_major_formatter(ticker.EngFormatter())
plt.show()
```

Using a subset of data from the current City of Chicago Salary Database, this chart displays the Mean Salaries for each Chicago Public Department within the data set. To provide more context on the how large each department is in relation to their salaries, the size of each department by employee number is indicated using differently colored bars. After looking at the data, the question I wanted to explore was how the salaries looked by department. 

I decided to use the mean of the salaries because I wanted to see how much each employee is paid on average.
