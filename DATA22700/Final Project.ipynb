```python
soybean_brazil_0113 = soybean_use.iloc[1210:1223]
soybean_brazil_0113.head()
```

```python
brazil_loss_merge = brazil_loss.merge(soybean_brazil_0113, how='inner')
brazil_loss_merge
```

```python
# Visualization 2 - Line graph - Merged data

pasture2 = brazil_loss_merge.pasture/1000000 #Millions
processed_feed = brazil_loss_merge.processed/1000000 #Millions
year2 = brazil_loss_merge.year

fig, ax1 = plt.subplots(figsize=(10,7))
ax2 = ax1.twinx()

line1 = ax1.plot(year2, pasture2, 'g-o', label = 'Loss of Forest Land by Pasture')
line2 = ax2.plot(year2, processed_feed, 'b-o', label='Processed Soy Production')

ax1.set_xlabel('Year', fontsize = 18)
ax1.set_ylabel('Loss of Forest Land for Pasture (Millions of Hectares)', color='g', fontsize = 16)
ax2.set_ylabel('Processed Soy (Millions of Tonnes)', color='b', fontsize = 18)
for label in ax1.yaxis.get_majorticklabels():
        label.set_fontsize(14)
for label in ax2.yaxis.get_majorticklabels():
        label.set_fontsize(14)
for label in ax1.xaxis.get_majorticklabels():
        label.set_fontsize(14)

plt.title('Loss of Forest Land vs. Production of Processed Soy in Brazil (2001-2013)', fontsize = 14)
plt.grid()
# Need to use fig.legend() when showing multiple lines and ylabels with ax.twinx()
fig.legend(loc=(0.63,0.42))
plt.show()

# Might need to change x and y tick sizes
```

```python
vegoil_brazil = vegetable_oil.iloc[14729:15425]
vegoil_brazil
```

```python
#defining lines

soybean = vegoil_brazil.loc[vegoil_brazil.crop_oil == 'Soybean']
soybean = soybean.loc[soybean.year < 2014]
print(soybean.shape)

# # sesame = vegoil_brazil.loc[vegoil_brazil.crop_oil == 'Sesame']

linseed = vegoil_brazil.loc[vegoil_brazil.crop_oil == 'Linseed']
linseed = linseed.loc[linseed.year < 2014]
print(linseed.shape)

palm = vegoil_brazil.loc[vegoil_brazil.crop_oil == 'Palm']
palm = palm.loc[palm.year < 2014]
print(palm.shape)

rapeseed = vegoil_brazil.loc[vegoil_brazil.crop_oil == 'Rapeseed']
rapeseed = rapeseed.loc[rapeseed.year < 2014]
print(rapeseed.shape)

groundnut = vegoil_brazil.loc[vegoil_brazil.crop_oil == 'Groundnut']
groundnut = groundnut.loc[groundnut.year < 2014]
print(groundnut.shape)

cottonseed = vegoil_brazil.loc[vegoil_brazil.crop_oil == 'Cottonseed']
cottonseed = cottonseed.loc[cottonseed.year < 2014]
print(cottonseed.shape)

coconut = vegoil_brazil.loc[vegoil_brazil.crop_oil == 'Coconut (copra)']
print(coconut.shape)

# safflower = vegoil_brazil.loc[vegoil_brazil.crop_oil == 'Safflower']
# print(safflower.shape)

sunflower = vegoil_brazil.loc[vegoil_brazil.crop_oil == 'Sunflower']
print(sunflower.shape)

maize = vegoil_brazil.loc[vegoil_brazil.crop_oil == 'Maize']
print(maize.shape)

# palm_kernal = vegoil_brazil.loc[vegoil_brazil.crop_oil == 'Palm kernal']

#define plot size
fig, ax = plt.subplots(figsize=(10,7))

#define colors to use in chart
color_map = ['red', 'steelblue', 'green', 'yellow', 'pink', 'purple', 'orange', 'brown', 'lightblue', 'magenta']
 
# Plot 4
plt.stackplot(soybean.year, np.log10(soybean.production), np.log10(linseed.production), np.log10(palm.production),
              np.log10(rapeseed.production), np.log10(groundnut.production), np.log10(cottonseed.production), 
              np.log10(coconut.production), np.log10(maize.production), np.log10(sunflower.production), 
              labels=['Soybean', 'Linseed', 'Palm', 'Rapeseed', 'Groundnut', 'Cottonseed', 'Coconut', 'Maize',
                      'Sunflower'],
              colors=color_map)

plt.title('Production of Plant-Based Oils in Brazil (1960-2013)', size=18)
plt.xlabel('Year', size=18)
plt.ylabel('Oil Production in Tonnes (log10)', size=18)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.legend(loc=(1,0.68), prop={'size': 16}, ncol=2)

plt.show()
```
