## 3. de Méré's problem ##

The Chevalier de Méré was a gambler, and found two propositions involving dice that had close to 50% probabilities:

**Event A**  In 4 rolls of the dice, six comes up at least once. 

**Event B**  In 24 rolls of the dice, six comes up at least twice.

He found out (by losing repeatedly) which of these was the winner, and he asked Blaise Pascal to look into the matter.


**Question 7.**
Write a function that simulates event A, returning True or False.

```python
dice = [1, 2, 3, 4, 5, 6]

def six_prob():
    """Calculates the probability that 6 comes up in 4 dice rolls"""
    result = np.random.choice(dice, 4)
    result_list = result.tolist()
    count = result_list.count(6)
    if count >= 1:
        return True
    else:
        return False

six_prob()
```

**Question 8.**
Write a function that simulates event B, returning True or False.

```python
dice = [1, 2, 3, 4, 5, 6]

def twentyfour_prob():
    """Calculates the probability that 6 comes up twice in 24 dice rolls"""
    result = np.random.choice(dice, 24)
    result_list = result.tolist()
    count = result_list.count(6)
    if count >= 2:
        return True
    else:
        return False

twentyfour_prob()
```

**Question 11.**
Write a function that simulates 1000 flips of a coin whose probability of heads is `heads_probability`

```python
def biased_flipper_1000(heads_probability):
    """Simulates 1000 coin tosses with heads_probability for the chance of landing on heads"""
    return np.random.binomial(1000, heads_probability, 1)

# The function returns a list of samples
#from a binomial distribution based on the inputted parameters when calling np.random.binomial(n, p, size).
#n = number of experiments, p = probability of success, size = number of randomly selected experiments

# Test
biased_heads = biased_flipper_1000(heads_probability)
biased_heads
```
