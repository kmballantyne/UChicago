## Part 1: numeric solution of the linear population model

Consider the linear difference equation below, where $N$ is the population vector and $r$ is the growth rate, expressed by the following iterative discrete-time model:

$$N(t+1)=rN(t)$$

This can represent a population that changes by an equal proportion every time step, e.g. if it grows by 5% a year, then $r=1.05$ and $t$ is time in years.

**Q1.1:** Let the population $N$ start with 700 individuals, with a constant mulitplicative growth rate of $r=0.8$. Use a for loop to propagate this dynamical systems for 20 timesteps, saving all the values into the array variable $N$. Then create another variable $t$ to represent time and plot $N$ as a function of time using the plt.plot commmand.

Self-check: the population after 20 iterations should be approximately 8.

```python
fig, ax = plt.subplots(figsize = (10,5))

steps = 20 #number of iterations
r = 0.8
pop = np.zeros(steps+1)
pop[0] = 700 # starting population N0
t = np.arange(steps+1)
print(t)

for i in range(steps):
    pop[i+1] = r*pop[i]

    
plt.plot(t, pop) # plot solution
plt.xlabel('Time', fontsize=20)
plt.ylabel('Population', fontsize=20)
title = "Population Size after 20 Iterations"
plt.title(title, fontsize=20) 
plt.show()

print(pop[20])
```

## Part 2: linear model with a constant additive term

Consider the modified linear growth model with a constant additive term $a$, and the other variables and parameters with the same meaning:

$$ N(t+1)=rN(t)+a $$

This may represent a population with a constant migration or death rate (e.g. in a controlled population like an aquarium or a nature preserve where a set number of hunting permits is issued.)

**Q2.1:** Let the population $N$ have the growth rate of $r = 0.95$ and the constant rate $a = 10$. Choose an initial valuer and compute the numeric solution of this model for for 100 timesteps, saving all the values into the vector variable $N$ and plot $N$ as a function of time using the plt.plot() commmand. Describe what kind of behavior you observe.  Try two other initial values, and plot the solutions in the same window with different colors.

Self-check: the population starting with $N(0)=400$ after 100 time steps should be about 201.

```python
fig, ax = plt.subplots(figsize = (10,5))

steps = 100 #number of iterations
r = 0.95
a = 10
pop = np.zeros(steps+1)
pop[0] = 400 # starting population N0
t = np.arange(steps+1)

for i in range(steps):
    pop[i+1] = (r*pop[i])+a

plt.plot(t, pop, label = '400')

print(pop[100])

pop[0] = 300 # initial value 1

for i in range(steps):
    pop[i+1] = (r*pop[i])+a

plt.plot(t, pop, label = '300')
print(pop[100])

pop[0] = 500 # initial value 2

for i in range(steps):
    pop[i+1] = (r*pop[i])+a
    
plt.plot(t, pop, label = '500')
print(pop[100])
    
plt.plot(t, pop) # plot solution
plt.xlabel('Time', fontsize=20)
plt.ylabel('Population', fontsize=20)
title = "Population Size after 20 Iterations"
plt.title(title, fontsize=20)
plt.legend()
plt.show()
```

The behavior we observe is an exponential decay for all three initial values. The initial values change where the starting population begins before it is iterated over time and observed. All three values end up iterating to a similar value of ~200-201 when t=100.

## Part 3: linear model with a random growth rate

Let us consider a more realistic model of a population with a linear growth rate than the deterministic (predictable) model with constant growth rate $r$: $N(t+1) = r N(t)$. In reality, no population has the exact same growth rate every cycle (whether it's a year or an hour) - there is always some unpredictable variation, called stochasticity. Let us assume that the population has the mean growth rate of 1.02 (or 2%) per year but this value fluctuates with a normal distribution, which means it's equally probable to be higher or lower than the mean. The big question is:

**Will the population with a fluctuating growth rate with mean value r grow faster, slower, or the same as the population with the constant growth rate r?**

Q3.1: Use your population model script from part 1 to simulate this situation, adding a *normal random number generator* with the mean of 1.02 a standard deviation of your choice to generate the growth rate r and iterate the population for 500 steps. Important hint: since we want to keep the growth rate positive, keep the standard deviation small enough that the random number is always positive. Generate the solution and plot its graph, and plot the solution for the deterministic population in the same figure. Can you make a guess about the answer of the question from this plot?

```python
fig, ax = plt.subplots(figsize = (10,5))

steps = 500
pop = np.zeros(steps+1)
pop[0] = 500 # starting population N0
t = np.arange(steps+1)

for i in range(steps):
    mu = 1.02
    sigma = 0.1
    num = 1
    r = np.random.normal(mu, sigma, num)
    pop[i+1] = r*pop[i]

plt.plot(t, pop, label = 'r = random') #r is random
    
r = 1.02

for i in range(steps):
    pop[i+1] = r*pop[i]

plt.plot(t, pop, label = 'r = 1.02') #r=1.02   

plt.xlabel('Time', fontsize=20)
plt.ylabel('Population', fontsize=20)
title = "Population Size after 500 Iterations"
plt.title(title, fontsize=20)
plt.legend
plt.show()
```

I think that the population with the fluctuating growth rate will grow slower than the population with the deterministic growth rate because the fluctuating growth rate can be less than 1 sometimes and cause the overall population to decline for a random number of time steps. The deterministic growth rate is more than 1 and will always increase after each time step
